```python
df['Gender']=df['Gender'].map({"Male":1,"Female":0})
```

```python
df['Workout_Type']=df['Workout_Type'].map({"Yoga":1,"HIIT":2,"Cardio":3,"Strength":4})
```

```python
df=df.drop_duplicates()
```

```python
x=df.drop('Workout_Type',axis=1)
x
```

```python
y=df['Workout_Type']
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python


from sklearn.preprocessing import StandardScaler


scalar=StandardScaler()

scalar.fit(x_train)

x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)

x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier

model=KNeighborsClassifier(n_neighbors=7)

model.fit(x_train,y_train)

pred=model.predict(x_test)
pred
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import RandomForestClassifier


knn=KNeighborsClassifier(n_neighbors=7)
nb=GaussianNB()
sv=SVC()
dt=DecisionTreeClassifier()

rf=RandomForestClassifier()

models = {
    "KNN": knn,
    "Naive Bayes": nb,
    "SVM": sv,
    "Decision Tree": dt,
    "Random Forest": rf
}


accuracy=[]
```

```python
for name, model_name in models.items():
    print("The model name is:", name)
    print("********************")
    
    model_name.fit(x_train, y_train)
    pred = model_name.predict(x_test)
    
    print(confusion_matrix(y_test, pred))
    print("Accuracy:", accuracy_score(y_test, pred))
    print(classification_report(y_test, pred))
    print("********************")
    
    accuracy.append(accuracy_score(y_test, pred))
```

```python
input_data = scalar.transform([[24,1,172,68,22.99,18.5,4,2,450,135,1.1,2.0,1,34.22]])
prediction = model.predict(input_data)
print(prediction)
```

```python
print("Train accuracy:", model.score(x_train, y_train))
print("Test accuracy:", model.score(x_test, y_test))
```

```python
plt.figure(figsize=(10,10))
s=plt.bar(models.keys(), accuracy,color='g')
plt.bar_label(s,fmt='%.2f%%')
plt.xlabel("Models")
plt.ylabel("Accuracy Score")
plt.title("Model Accuracy Comparison")
plt.ylim(0, 1)   # accuracy range
plt.show()
```
